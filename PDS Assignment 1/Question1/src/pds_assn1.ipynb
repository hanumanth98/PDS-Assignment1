```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('rawdata_frality.csv')
```

```python
df.describe()
```

```python
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
frailty_groups = df.groupby('Frailty')
for var in ['Height', 'Weight', 'Age', 'Grip strength']:
    print(f"\nMean {var} by Frailty Group:")
    print(frailty_groups[var].mean())
```

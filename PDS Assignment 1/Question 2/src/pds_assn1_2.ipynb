```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df
```

```python
df.describe()
```

```python
monitor = df.isnull()
df = df.dropna(subset = monitor.columns, how = 'all')
df
```

```python
male_scores = df[df['gender'] == 'male']['reading score']
female_scores = df[df['gender'] == 'female']['reading score']

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(range(len(male_scores)), male_scores, label='Male', c='blue', marker='o')
plt.scatter(range(len(female_scores)), female_scores, label='Female', c='red', marker='x')

# Adding labels and title
plt.xlabel('Student')
plt.ylabel('Scores')
plt.title('Student Scores by Gender')
plt.legend()

# Show the plot
plt.show()
```

```python
df[['math score', 'reading score', 'writing score']].plot(kind='hist', alpha=0.7, bins=10)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math, Reading, and Writing Scores')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='gender', y='math score', palette='Set2')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Math Scores by Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='test preparation course', palette='Set3')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.title('Count of Students with/without Test Preparation Course')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='parental level of education', palette='viridis')
plt.xlabel('Count')
plt.ylabel('Parental Level of Education')
plt.title('Distribution of Parental Level of Education')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='race/ethnicity', y='reading score', data=df, palette='Paired')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Reading Score')
plt.title('Reading Scores by Race/Ethnicity')
```

```python
sns.pairplot(df[['math score', 'reading score', 'writing score']], height=2)
plt.suptitle('Pairplot of Math, Reading, and Writing Scores')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Boxplot of Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df)
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Math Scores by Gender')
```
